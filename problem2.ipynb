## Problem 2: Wrangling the Billboard Top 100

```python
import pandas as pd
```

```python
billboard_data = pd.read_csv("data/billboard.csv")
```

### Part A

```python
billboard_data[["performer","song"]].value_counts()[:10].reset_index().rename(columns={0:'count'})
```

This table shows the top songs on the billboard 100 since 1958 till week 22 of 2021.

We see that the top 10 songs have been popular for longer than 1 year

### Part B

```python
billboard_data.drop_duplicates(["year","performer","song"]).groupby("year")["week_position"].count()[1:-1].plot(xlabel="Years", ylabel="Unique Songs", title="Unique Songs per Year in Billboard Top 100")
```

The plot shows the amount of unique songs for each year from 1959 to 2020. This allows us to understand the musical diversity in different years.\
During 1985 to 2009 there was a low diversity on music on the top 100, with 2001 being the lowest diverse year wit 387 different songs, but it has been increasing during the last years. 

### Part C

```python
ten_week_hit = billboard_data[["performer","song"]].value_counts().reset_index().rename(columns={0:'count'})
ten_week_hit = ten_week_hit[ ten_week_hit["count"] >= 10]

top_ten_artist = ten_week_hit.groupby("performer")["count"].count().sort_values(ascending= False)
top_ten_artist[ top_ten_artist.values >= 30].plot(kind="bar", title="Performers with more 10-week hit on the Billboard Top 100")
```

Elton John is the top performer in the last 60 years, with more than 50 ten-week hits, having a difference of 8 songs to Madonna the runner up. This is the highest difference between the top artists
